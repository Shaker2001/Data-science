# linear regression Gradient Descent in Python


```python
data_url = "https://raw.githubusercontent.com/chasinginfinity/ml-from-scratch/master/02%20Linear%20Regression%20using%20Gradient%20Descent/data.csv"
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# we will use Customizing Matplotlib with style sheets and rcParams

```python
plt.rcParams['figure.figsize'] = (10.0 , 10.0)
dt = pd.read_csv(data_url)
X = dt.iloc[:,0]
Y = dt.iloc[:,1]
plt.scatter(X,Y)
plt.show()
```

#Building the model linear regression

```python
m= 0 
c = 0
L = .0001 # that's learning Rate
steps = 1000 #the recommended number of iterations to perform the algo
n = len(X)
for i in range(steps):
    y_calc = m*X + c
    Dev_m = (-2/n) * sum(X*(Y-y_calc))
    Dev_c = (-2/n) * sum((Y-y_calc))
    m = m- L * Dev_m # updating m 
    c = c- L * Dev_c #updating c
print(m,c)
```

```python
#so we can make predictions 
y_predict = m* X + c
plt.scatter(X,Y)
plt.plot([min(X), max(X)],[min(y_predict),max(y_predict)],color='red')
plt.show()
```
